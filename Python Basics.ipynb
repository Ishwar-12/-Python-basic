2. What is an interpreter in Python?

- An interpreter in Python is a program that reads and executes Python code line by line.
  Unlike compiled languages like C or Java, where the entire code is converted into machine code before execution, Python uses an interpreter to:
  Translate one line at a time into machine code
  Immediately execute each line after translating
  Report errors as they occur, making debugging easier
  This makes Python more flexible and interactive, which is especially useful for beginners and rapid development.

5. What is mutability in Python?
- Mutability in Python refers to an object’s ability to be changed after creation.

- Mutable objects like: list, dict, set, bytearray,etc can be modified.

- Immutable objects like: int, float, str, tuple, bool, frozenset, etc cannot be modified.

- Mutability impacts memory management, determines whether to copy or reference objects, and affects function behavior, especially with default arguments.









6. Why are lists mutable, but tuples are immutable?
- Lists are mutable because they'redesigned for flexibility and changing data, supporting operations like append( ) and item assignment. Tuples are immutable to ensure safety, better performance, and efficient memory use. Their fixed nature makes them faster and safe to use as dictionary keys or set elements.

7. What is the difference between “==” and “is” operators in Python?
- (==)	Equality operator	If values are equal	`a == b`
- (is)	Identity operator	If both refer to the same object in memory	`a is b`
- We use` == `for comparing values, and `is` for checking object identity (like `is None`).

```python
#Ex
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)  #  True — same content
print(a is b)  #  False — different objects in memory
```

8. What are logical operators in Python?
-  Logical operators in Python `and`, `or`, and - `not` are used to combine or modify conditional statements, returning True or False based on logic:

- `and`  True if both conditions are true

- `or`  True if at least one condition is true

- `not`  Inverts the Boolean value

- They help control program flow by evaluating logical expressions.

```python
#Ex
x = 10
y = 5

print(x > 5 and y < 10)   # True (both conditions are True)
print(x < 5 or y < 10)    # True (one condition is True)
print(not(x == 10))       # False (x == 10 is True, so not makes it False)
```

12. How does the elif statement work?
- The `elif` statement allows checking multiple conditions after an if. Python evaluates them in order, and only the first True condition is executed. If none are true, the else block runs.
- Python checks the if condition first.

- If it's False, it moves to elif and checks that condition.
It continues checking any additional elif blocks in order.
If none are True, it runs the else block.

```python
x = 30

if x < 10:
    print("Less than 10")
elif x < 30:
    print("Between 10 and 30")
else:
    print("30 or more")
```

13. What is the difference between for and while loops?
- `for` loops are used when the number of iterations is known and you're looping over a sequence.
- `while` loops are used when the number of iterations is unknown and controlled by a condition.

```python
#Ex
#for loop

for i in range(3):
    print(i)
# Output: 0, 1, 2
```

```python
#while loop

i = 0
while i < 3:
    print(i)
    i += 1
# Output: 0, 1, 2
```

14. Describe a scenario where a while loop is more suitable than a for loop.
- Scenario Where a `while` Loop is More Suitable Than a `for` Loop:
-- User Input Until Correct Response

- When you want to keep asking a user for input until they enter a valid response, a while loop is more appropriate because you don’t know how many attempts they’ll need.
- We don’t know how many tries the user will need.
- The loop runs as long as the condition is unmet, which is ideal for input validation, waiting for a signal, or retry logic.

```python
#5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each.
#Ans:

# Creating a list and a tuple
my_list = [10, 20, 30]
my_tuple = (10, 20, 30)

# Modifying the list (mutable)
print("Original list:", my_list)
my_list[1] = 99
print("Modified list:", my_list)

# Attempting to modify the tuple (immutable)
print("Original tuple:", my_tuple)
try:
    my_tuple[1] = 99  # This will raise an error
except TypeError as e:
    print("Error when trying to modify tuple:", e)
```

```python
#6. Write a function to demonstrate the behavior of mutable and immutable arguments.
#Ans:

# Function to modify both mutable and immutable arguments
def modify_args(num, my_list):
    num += 10          # This won't affect the original variable (immutable)
    my_list.append(99) # This will change the original list (mutable)
    print("Inside function:")
    print("num =", num)
    print("my_list =", my_list)

# Immutable argument (int)
x = 5

# Mutable argument (list)
lst = [1, 2, 3]

# Call the function
modify_args(x, lst)

print("\nOutside function:")
print("x =", x)         # Unchanged
print("lst =", lst)     # Changed

#num is an immutable integer >> changes inside the function don't affect the original.

#my_list is a mutable list >> changes inside the function do affect the original.
```

```python
#9. Write a Python program to convert user input from string to integer, float, and boolean types.
#Ans:

# Get user input as a string
user_input = ("45")

# Convert to integer
try:
    int_value = int(user_input)
    print("Integer:", int_value)
except ValueError:
    print("Cannot convert to Integer")

# Convert to float
try:
    float_value = float(user_input)
    print("Float:", float_value)
except ValueError:
    print("Cannot convert to Float")

# Convert to boolean
# In Python, empty strings are False, non-empty strings are True
bool_value = bool(user_input)
print("Boolean:", bool_value)
```

```python
#10. Write code to demonstrate type casting with list elements.
#Ans:

# List of strings representing numbers
str_list = ["10", "20", "30", "40"]

# Convert string elements to integers
int_list = [int(item) for item in str_list]
print("Converted to integers:", int_list)

# Convert integer elements to floats
float_list = [float(item) for item in int_list]
print("Converted to floats:", float_list)

# Convert float elements to strings
str_again_list = [str(item) for item in float_list]
print("Converted back to strings:", str_again_list)

# Convert non-empty strings to boolean
bool_list = [bool(item) for item in str_list]
print("Converted to booleans:", bool_list)
```

```python
#14. Write a program to reverse a string using a while loop.
#Ans:

# Get user input
text = ("Ishwar")

# Initialize index and reversed string
index = len(text) - 1
reversed_text = ""

# Reverse using while loop
while index >= 0:
    reversed_text += text[index]
    index -= 1

# Display the result
print("Reversed string:", reversed_text)
```

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
#Ans:

num = int("12")

# Initialize
factorial = 1
i = 1

# calculate factorial
if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    while i <= num:
        factorial *= i
        i += 1
    print(f"Factorial of {num} is: {factorial}")
```
